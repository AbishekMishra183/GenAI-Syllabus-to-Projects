# joining numpy array

```python
a=np.arange(10,20)
b=np.arange(20,30)
c=np.arange(30,40)
d=np.arange(40,50)
e=np.arange(50,60)
```

```python
a=np.arange(21,30,1).reshape(3,3)
a
```
