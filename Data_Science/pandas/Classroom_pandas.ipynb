# Series objective in the 1 d labelled  array

```python
import pandas as pd
```

```python
s1=pd.Series([1,2,3,4,5])
```

```python
s1
```

```python
type(s1)
```

# series value can be indexed  and change to its value

```python
s2=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
```

```python
s2
```

# we can also create a a series obj from the dict 

```python
s3=pd.Series({'a':10,'b':20,'c':30})
```

```python
s3
```

# We can change the index position

```python
pd.Series({'a':10,'b':20,'c':30}, index=['b','c','d','a'])
```
